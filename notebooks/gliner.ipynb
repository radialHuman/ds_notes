# General

```python
import pandas as pd
```

```python
from gliner import GLiNER

# Initialize GLiNER with the base model
model = GLiNER.from_pretrained("urchade/gliner_medium-v2.1")
```

```python
# Sample text for entity prediction
text = """
i have a hyundai i10 but something is wrong with the seats and the secret it looks like it is not working.
airtel mobile connection is not working properly
there is a huge corruption issue with the ministry of culture
remote setup sports we can t avoid people drinking water
canara bank is not allowed inside as a do not have an id card
my bajaj pulsar is one of the worst bike ever i am not going to buy any more bajaj products bye bye bajaj
the roads are not good as it is full of traffic jam even in the cleaners
i bought my iphone today is back and now its not working
we need more cell phone towers in this area because i am not able to make calls inside and outside
there are too many bikes on the road these days
my renault duster tyre servicing wasnot good
the streets and noisy and no one can concentrate on anything they are doing over there because the traffic jam
the flights are not in proper shape and any maintenance
my life sucks
the indian bank is not allowing me inside
"""

# Labels for entity prediction
# Most GLiNER models should work best when entity types are in lower case or title case
labels = ["car company", "government department", "internet provider", "workplace", "life", "traffic", "bike", "phone", "connection", "bank", "airplane"]

# Perform entity prediction
entities = model.predict_entities(text, labels, threshold=0.2)

# Display predicted entities and their labels
output = {}
for entity in entities:
  output[entity["text"]] = entity["label"]
output
```

```python
# Sample text for entity prediction
text = """
i have a hyundai i10 but something is wrong with the seats and the secret it looks like it is not working.
airtel mobile connection is not working properly
there is a huge corruption issue with the ministry of culture
remote setup sports we can t avoid people drinking water
canara bank is not allowed inside as a do not have an id card
my bajaj pulsar is one of the worst bike ever i am not going to buy any more bajaj products bye bye bajaj
the roads are not good as it is full of traffic jam even in the cleaners
i bought my iphone today is back and now its not working
we need more cell phone towers in this area because i am not able to make calls inside and outside
there are too many bikes on the road these days
my renault duster tyre servicing wasnot good
the streets and noisy and no one can concentrate on anything they are doing over there because the traffic jam
the flights are not in proper shape and any maintenance
my life sucks
the indian bank is not allowing me inside
"""

# Labels for entity prediction
# Most GLiNER models should work best when entity types are in lower case or title case
labels = ["car company", "government department", "internet provider", "workplace", "life", "traffic", "bike", "phone", "connection", "bank", "airplane"]

# Perform entity prediction

# Display predicted entities and their labels
output = {}
for t in text.split("\n"):
  output[t] = []
  entities = model.predict_entities(t, labels, threshold=0.6)
  for entity in entities:
    output[t].append({entity["text"]: entity["label"]})
pd.DataFrame.from_dict(output, orient='index')
```

```python
# Initialize GLiNER with the base model
model = GLiNER.from_pretrained("urchade/gliner_large_bio-v0.1")
```

```python
# Sample text for entity prediction
text = """
after taking the pill, I developed red rashes around my neck
there were air bubbles in the syringe
i was not able to push the pen of the injection
the doctor said the nurse will help me with all my queries regarding the pill
i was not able to sleep after taking the medicine
"""
```

```python
# Labels for entity prediction
# Most GLiNER models should work best when entity types are in lower case or title case
labels = ["discomfort", "medical device"]

# Perform entity prediction
entities = model.predict_entities(text, labels, threshold=0.5)

# Display predicted entities and their labels
for entity in entities:
    print(entity["text"], "=>", entity["label"])
```
