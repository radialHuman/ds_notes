```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
complain = "after using your broken syringe I had blurry vision and pain in the area"

prompt = """Keep the following rules in mind
1. If it does not feel like a complaint then return "NA"
2. If there is a complain about side effect from a medicine or medical device return "AE"
3. If there is a product complaint regarding a medical device return "PC"
4. If has complaints about both the product and the side effect return "Both"
5. Reevaluate and reson deeply to no miss any of the complains.
you have to identify if the following sentence has any information about a side effect from medicine or a defect in medical device

<SENTENCE>
{}
</SENTENCE>
""".format(complain)
messages = [{"role": "You are a medical expert", "content": ""},{"role": "user", "content": prompt}]
```

```python
# 1.5 QWEN
```

```python
model_name_1_5 = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"
model_1_5 = AutoModelForCausalLM.from_pretrained( model_name_1_5, torch_dtype="auto", device_map="auto")
tokenizer_1_5 = AutoTokenizer.from_pretrained(model_name_1_5)
```

```python
%%time
text_1_5 = tokenizer_1_5.apply_chat_template(messages, tokenize=False,add_generation_prompt=True)
model_inputs_1_5 = tokenizer_1_5([text_1_5], return_tensors="pt").to(model_1_5.device)
generated_ids_1_5 = model_1_5.generate(**model_inputs_1_5, temperature=0.1, max_new_tokens=5512)
generated_ids_1_5 = [output_ids_1_5[len(input_ids_1_5):] for input_ids_1_5, output_ids_1_5 in zip(model_inputs_1_5.input_ids, generated_ids_1_5)]
response_1_5 = tokenizer_1_5.batch_decode(generated_ids_1_5, skip_special_tokens=True)[0]
print(response_1_5)
```
