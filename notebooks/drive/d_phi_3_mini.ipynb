```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
torch.random.manual_seed(0)
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}
```

```python
%%time

complain = '''
'52 | I already complaint against ucoin Google review team, now I here to see many people complaint against ucoin. This company say to you did mistake, so you pay full amount to unfreeze your  account, this type of complaints are so many.
'''
messages = [
    {"role": "system", "content": "You are a language expert who knows everything about complains and products and organization names"},
    {"role": "user", "content": "From the following sentence can you extract 1. the brand name or the name of the company or organization \n 2. Extract the exact product the complain is against  : <sentence>{}</sentence> \n 3. Is the issue going on for a long time yes or no? \n 4. Is it some sort of emergency yes or no?".format(complain)},
    # {"role": "assistant", "content": "Sure! Here are some ways to eat bananas and dragonfruits together: 1. Banana and dragonfruit smoothie: Blend bananas and dragonfruits together with some milk and honey. 2. Banana and dragonfruit salad: Mix sliced bananas and dragonfruits together with some lemon juice and honey."},
    # {"role": "user", "content": "What about solving an 2x + 3 = 7 equation?"},
]
output = pipe(messages, **generation_args)
print(output[0]['generated_text'])
```
